```python
import glob
import os
import numpy as np
import scendict as sc
from os.path import exists
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams.update({'font.size': 14})
font = {'family' : 'normal',
'weight' : 'bold',
'size'   : 14}

runhorse = True
```

```python
def make_symlink_if_exists(src,dst,yr,var):
    if exists(src) == False:
        print(f'year {yr}, {var} original file doesnt exist')
    if exists(src):
        try:
            os.symlink(src, dst)
        except:
             print(f'year {yr}, {var} original file exists but symlink not made')

bdir = '/gpfs/home/mep22dku/scratch/MET_soft/'

scenlist = ['1A','1B','2A','2B','3A','3B']
if runhorse:
    for s in range(0,6):
        tscen = scenlist[s]
        print(f'making symlinks for scenario {tscen}')
        for yr in range(1940,2101):
            if yr < 2015:
                ldir = sc.scen[tscen]['hist_dir']
            else:
                ldir = sc.scen[tscen]['fut_dir']
            tdir = f'{bdir}{ldir}' 

            #tair
            src = f'{tdir}/bulk_{yr}_16_met_daily.nc'; var = 'tair'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_tair10m{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/bulk_{yr}_17_met_daily.nc'; var = 'qair'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_qair10m{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/bulk_{yr}_18_met_daily.nc'; var = 'sw_down'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_sw_down{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/bulk_{yr}_19_met_daily.nc'; var = 'lw_down'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_lw_down{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/bulk_{yr}_20_met_daily.nc'; var = 'precip'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_precip{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/bulk_{yr}_21_met_daily.nc'; var = 'snow'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_snow{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/taux_1d_{yr}_daily.nc'; var = 'taux'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_taux{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

            src = f'{tdir}/tauy_1d_{yr}_daily.nc'; var = 'tauy'
            dst = f'/gpfs/home/mep22dku/scratch/MET_forcing/scen_{tscen}/MetOffice_tauy{yr}.nc'
            make_symlink_if_exists(src,dst,yr,var)

```
